```
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

```
def insideoroutside(x,y,R):
  dx = abs(x)
  dy = abs(y)
  if(dx + dy <= R): 
      return True
  if(dx > R):
      return False
  if(dy > R):
      return False
  if(dx**2 + dy**2 <= R**2):
      return True
  else:
      return False
```

```
invals = 0
outvals = 0
sample = 100
xvals = [0 for i in range(sample)]
yvals = [0 for i in range(sample)]
for i in range(0,sample):
  xvals[i] = (rd.uniform(-1,1))
  yvals[i] = (rd.uniform(-1,1))
  if(insideoroutside(xvals[i],yvals[i],1) is True):
    invals += 1
  else:
    outvals += 1
circle1 = plt.Circle((0,0), 1, color='b', fill=False)
fig,ax = plt.subplots()
ax.add_patch(circle1)
plt.plot([xvals],[yvals],'ro',markersize=1)
plt.axis([-1,1,-1,1])
plt.show
print(invals)
print(" points are inside the circle")
print(outvals)
print(" points are outside the circle")
if(outvals is 0):
  print("Pi cannot be calculated since no points are outside the circle. Try a larger number of points")
else:
  estpi = invals*4/sample
  print(estpi)
  print(" is the estimated value of Pi")
```
